```
#Predict the feature ‘y’ using 'BMI'
# Model’s coefficients (slope, y-intercept)
# The Linear Regressor Model (graph) plotting
# The MSE (Mean Square Error)
```

```
from sklearn import preprocessing
X = df[['BMI']]
y = df[['Y']]
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=1/4 ,random_state =0)
```

```
X_train.shape
```

```
ln_model = LinearRegression()
ln_model.fit(X_train,y_train) #to find the values of m and b such that y=mx+b describes a line that fits our observed data well.
yHat_test = ln_model.predict(X_test) #predicted value
yHat_train = ln_model.predict(X_train) #predicted value

print('Intercept:',ln_model.intercept_) #get the model intercept
print()

print('Slope:' ,ln_model.coef_) #get the model co-efficient
print()

print('Accuracy score of training set:',ln_model.score(X_train,y_train))
print()
print('Accuracy score of test set:',ln_model.score(X_test,y_test))
print()
```

```
from sklearn import metrics
MAE = round(metrics.mean_absolute_error(y_test,yHat_test) ,2)
MSE = round(metrics.mean_squared_error(y_test,yHat_test) ,2)
RMSE = round(np.sqrt(metrics.mean_squared_error(y_test,yHat_test)) ,2)
RS= round(metrics.r2_score(y_test,yHat_test),2)

print('MAE:',MAE) 
print('MSE:',MSE)
print('RMSE:',RMSE)
print('R^2 score:',RS)
```

```
plt.figure(figsize =(10,7))
plt.scatter(y_test,yHat_test ,color='r',alpha =0.8)
plt.plot(y_test,y_test,color ='k',label = 'y=y')
plt.title('y vs predicted y')
plt.xlabel('y')
plt.ylabel('predicted y')
plt.legend()
plt.show()
```
